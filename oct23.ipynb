```python
import numpy as np
import matplotlib.pyplot as plt

# Define the velocity function
def velocity_function(x, y):
    u = 10.0 * np.exp(-((x - 50000.0)**2 + (y - 15000.0)**2) / (2 * 10000.0**2))
    v = 5.0 * np.sin(np.pi * x / 100000.0) * np.cos(np.pi * y / 30000.0)
    return u, v

# Generate a grid of points
x = np.linspace(0, 100000, 100)
y = np.linspace(0, 30000, 100)
X, Y = np.meshgrid(x, y)

# Compute the velocity components
U, V = velocity_function(X, Y)

# Plot the velocity field
plt.figure(figsize=(10, 6))
plt.quiver(X, Y, U, V, scale=50)
plt.title('Velocity Field')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()

```
